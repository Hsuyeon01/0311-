### 다변량 그래프 그리기 ###

- 3개 이상의 변수를 사용한 그래프를 '다변량 그래프'라고 부름

**다변량 그래프가 필요한 이유**

- 3개 이상의 변수를 사용하는 다변량 그래프는 그래프 요소를 추가하여 그래프를 표현해야 함 (잠의 크기, 점의 색상 등)

**다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프**

1. 산점도 그래프에 성별을 새 변수로 추가, Female Male과 같은 문자열은 산점도 그래프의 색상을 지정한느 값으로 사용할 수 없기 때문에 0,1과 같은 정수를 사용해야 함, 따라서 각 문자열을 정수로 치환하는 함수를 만들어야 함

```python
#여성인 경우에 0을 반환하고, 남성인 경우에 1을 반환하는 함수

def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

2. recode_sex 메서드가 반환한 값(0,1)을 데이터 프레임에 추가하면 됨, 다음은 recode_sex 함수가 반환한 값(0,1)을 데이터 프레임 sex_color 열을 새로 생성하여 추가한 것, 이때 sex열에 recode_sex 함수를 브로드캐스팅하기 위하여 apply 메서드를 사용

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips.head()
```

3. 테이블당 인원 수는 점의 크기로 표현하면 적당, 다음은 scatter 메서드에 s,c 인잣값으로 테이블당 인원 수와 성별의 치환값을 전달한 것 (s는 점의 크기를, c는 점의 색상을 의미, alpha로 투명도 조절)

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,
    c = tips['sex_color'], alpha=0.5)

axes1.set_title('Total Bill v Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

2. 산점도 그래프와 히스토그램을 한번에 그려주는 jointplot 메서드, x,y 인자에 원하는 열 이름을 지정하고 data 인잣값으로 데이터프레임을 지정하면 그래프를 그릴 수 있음

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

3. 산점도 그래프는 점이 겹쳐 보일 경우 점을 구분하기 어렵다는 단점이 있음 - 육각 그래프를 사용하면 됨, 육각그래프는 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프로 특정 데이터의 개수가 많아지면 점점 진한 새으로 표현됨, 육각 그래프는 jointplot 메서드를 사용하되 kind 인잣값을 hex로 지정하면 됨 

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

4. 이차원 밀집도 그리기

- 이차원 밀집도는 kdeplot 메서드로 그릴 수 있음

```python
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                data2 = tips['tip'],
                shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
fig = sns.pairplot(tips)
```

9. 관계 그래프는 중복된 정보가 표현된다는 단점이 있음, 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체

- map_upper 메서드는 대각선을 기준으로 위쪽에 그릴 그래프를 지정
- map_lower 메서드는 대각선을 기준으로 아래쪽에 그릴 그래프를 지정
- map_diag는 대각선을 중심으로 그래프를 그림

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()
```

## seaborn 라이브러리로 그래프 스타일 설정하기 ##

seaborn 라이브러리로 만든 그래프 스타일은 set_style 메서드를 사용하여 바꿀 수 있음, 메서드를 실행한 이후의 그래프가 모두 지정한 스타일로 바뀜

**그래프에 스타일 적용하기**

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

2. whitegrid 스타일 설정 (그래프의 배경에 가로 줄이 생김)

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

3. for문을 이용하여 모든 스타일을 하나씩 적용한 그래프

```python
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax=fig.add_subplot(2,3,plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)
        
fig.tight_layout()
```
